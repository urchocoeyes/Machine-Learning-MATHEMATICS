Let $x_1, \dots, x_n \sim U[0,1]$. Sort them in the nondecreasing order:

$$
    x_{(1)} \leqslant x_{(2)} \leqslant \ldots \leqslant x_{(k)} \leqslant\ldots\leqslant x_{(n)}.

The $k$-th element $x_{(k)}$ in this sequence is called **$k$-th order statistics**. Show that $x_{(k)} \sim \mathrm{Beta}(k, n+1 -k)$. Do it in two ways:

* **experimentally**: generate a random matrix of shape $N \times n$ (try $N = 1000, 10000, 100000$), calculate $k$-th order statistics for each row and plot historgram of these values along with pdf of beta distribution. (**1 point**)

* **theoretically**: present a mathematical proof of this statement (**1 point**)

### MY SOLUTION
Theoretical proof: Cumulative distribution function of $k$-th order statistic $X_{(k)}$:
$$
F_{x_{k}}(x) = P(X_{(k)} <= x) 
\\
= P(\text{at least k of the } X_{i}\text{'s are} <= x) \quad \text{where } X_i \text{ is i-th  element of X}
\\
= \sum_{j=k}^n P(\text{exactly j of n } X_{i}\text{'s are} <=x) \quad \text{Summation means "OR" operator}
$$
We see the sentence "exactly $j$ of $n$ $X_{i}$", this is Binomial probability:
$$
= \sum_{j=k}^n \binom{n}{j}[F_{X}(x)]^{j} [1 - F_{X}(x)]^{n-j}
$$
In case of **Uniform distribution with interval [0, 1]**:
$$
F_{X}(x) = \begin{cases}0, \quad x <= 0
\\
\frac{(x-0)}{(1-0)} = x, \quad 0<x<=1
\\
1, \quad x > 1
\end{cases}
$$
Recognize that $F_{x_{k}}(x)$ matches the CDF of beta distribution with parameters $k$ and $n+1-k$:
$$
F_{\text{Beta}(k, n+1-k)} = \int_0^1 t^{k-1}(1-t)^{(n+1-k)-1}dt
$$
This concludes the mathematical proof showing that the $k-th$ order statistic $x_{(k)}$ follows beta distribution with parameters $k$ and $n+1-k$ .
$$$$
****
Experimentally with code:

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Setting random matrix' size
N_values = [1000, 10000, 100000] # Samples
n = 13 # in each sample we will have n random numbers
fig, axes = plt.subplots(len(N_values), figsize=(10, 8), sharex=True) # returns three figures (3 figures will be stored in 'fig' variable, each figure corresponds to each value in N_values, all figures will have the same 'x' axes)

k = 7 # 's order statistics

for i, N in enumerate(N_values):
    X = np.random.uniform(0, 1, size=(N, n)) # generated matrix with size N x n with random uniformly distributed values 
    order_statistics = np.sort(X, axis=1)[:, k-1]  # Selecting the kth order statistics, up to k-1, since will be selected by indices.
    
    # Plot histogram
    axes[i].hist(order_statistics, bins=25, density=True, alpha=0.9, label='Empirical')

    # Plot beta distribution
    beta_pdf = beta.pdf(order_statistics, k, n + 1 - k)
    axes[i].plot(order_statistics, beta_pdf, 'y-', lw=2, label='Beta PDF')
    axes[i].set_title(f'N = {N}')
    axes[i].legend()

plt.xlabel('k-th order statistics')
plt.ylabel('Density')
plt.show()
````

## Task 4.3 (1 point)

Calculate entropy of $\mathcal N(\mu, \sigma^2)$. Plot the graph of this entropy as function of $\sigma$. What is the limit of this function as $\sigma \to +0$? $\sigma \to +\infty$?


### My solution
Given probability density function(pdf) $p_{\xi}(x)$ , differential entropy of continuous random variable $\xi$ is calculated by the formula:
$$
\mathbb H \xi = -\int p_{\xi}(x)\log p_{\xi}(x)dx \quad (1)
$$

The logarithm base can be any number greater than 1. In case of normal distribution, it is convenient to use **natural** logarithm(with base $e$).

For normal distribution $N(\mu, \sigma ^2)$, the probablity density function is:
$$
p(x| \mu, \sigma^2) = \frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}  \quad (2)
$$

To find entropy of normal distribution, we need to substitude (2) into (1).
$$
    \mathbb H = -\int p( x|\mu, \sigma^2) \ln (\frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}})dx \quad (3)
$$
Logarithmic fromulas:
$$
\log _b(mn) = \log_b(m) + \log_b(n)
\\
\log _b (b) = 1
\\
\log_b (m^n) = n\log_b m
$$
Trying to find $\ln (\frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}})$ with logarithmic formulas, in expression (3):
$$
\ln (\frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}})
\\
= \ln (\frac{1}{\sqrt{2\pi \sigma^2}}) + \ln(e^{-\frac{(x-\mu)^2}{2\sigma^2}})
\\
= -\frac{1}{2} \ln (2\pi \sigma^2) + (-\frac{(x-\mu)^2}{2\sigma^2})
\\
= -\frac{1}{2} \ln (2\pi \sigma^2) -\frac{1}{2\sigma^2}(x - \mu)^2 \quad (4)
$$

I used logarithm properties:

$$
\ln(e^{-\frac{(x-\mu)^2}{2\sigma^2}}) = \log _e(e^{-\frac{(x-\mu)^2}{2\sigma^2}}) = -\frac{(x-\mu)^2}{2\sigma^2}
\\
\ln (\frac{1}{\sqrt{2\pi \sigma^2}}) = \ln(\sqrt{2\pi \sigma^2})^{-1} = \ln(({2\pi \sigma^2})^{\frac{1}{2}})^{-1} = \ln ({2\pi \sigma^2})^{-\frac{1}{2}} = -\frac {1}{2}(\ln (2\pi \sigma^2))
$$

Substituting (4)-th expression to (3)-rd expression, i.e, I've found logarithmic part of normal distribution's entropy, and I'm going to substitude that resulting logarithmic expression to entropy expression:

$$
\mathbb H = -\int p( x|\mu, \sigma^2) \ln (\frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}})dx 
\\
\ln (\frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}})
= -\frac{1}{2} \ln (2\pi \sigma^2) -\frac{1}{2\sigma^2}(x - \mu)^2 
\\
\mathbb H = -\int_{-\infty}^{+\infty} p( x|\mu, \sigma^2) (-\frac{1}{2} \ln (2\pi \sigma^2) -\frac{1}{2\sigma^2}(x - \mu)^2 ) dx
\\
= -\int_{-\infty}^{+\infty} p( x|\mu, \sigma^2) \cdot (- (\frac{1}{2} \ln (2\pi \sigma^2) + \frac{1}{2\sigma^2}(x - \mu)^2 )) dx \quad {\text{obtained 2 minus signs,    }}
\\
= \int_{-\infty}^{+\infty} p( x|\mu, \sigma^2)  (\frac{1}{2} \ln (2\pi \sigma^2) + \frac{1}{2\sigma^2}(x - \mu)^2 )dx 
\\ 
= \frac{1}{2}\ln (2\pi \sigma^2) \int_{-\infty}^{+\infty} p( x|\mu, \sigma^2) dx + \frac{1}{2\sigma^2} \int_{-\infty}^{+\infty} (x - \mu)^2 p( x|\mu, \sigma^2) dx 
$$
Note that:
$$
\int_{-\infty}^{+\infty} p( x|\mu, \sigma^2) dx = 1
\\
\int_{-\infty}^{+\infty} (x - \mu)^2 p( x|\mu, \sigma^2) dx = \sigma^2 \quad {\text {=variance(dispersion)}}
$$
Dispersion is sigma squared: $D = \sigma^2$
Finally:
$$
\mathbb H = \frac{1}{2}\ln (2\pi \sigma^2) \cdot 1 + \frac{1}{2\sigma^2} \cdot \sigma^2
\\
= \frac{1}{2}(\ln (2\pi \sigma^2) + 1)
$$
I can expand it, since $\ln e = 1$:
$$
\mathbb H = \frac{1}{2}(\ln (2\pi \sigma^2) + \ln e )
\\
= (\ln (2\pi \sigma^2)+ \ln e )^{\frac{1}{2}} 
\\
= \ln (\sqrt{2\pi \sigma^2 \cdot e})
\\
= \ln (\sigma \sqrt{2\pi \cdot e})
$$

# Plot the graph of this entropy as function of 

````python
# Plotting the graph of entropy as function of sigma 
import numpy as np
import matplotlib.pyplot as plt

def entropy_normal_distribution(sigma): 
    # entropy of normal distribution depends on only sigma, not mean, because, we proved that above, we have no mean in result.
    return 0.5 * np.log(2 * np.pi * np.exp(1) * sigma**2) # np.log is by default --> natural logarithm with base "e", np.pi --> 3.14, np.exp(1) --> e^1

# Generate values of sigma, close to 0 and different from 0,
# (cannot use np.inf, because range is limited in linspace, also cannot use 0, because, 0 is answer for "no power" from 
# point of view of LOGARITHMS (in function entropy_normal_distribution, if we will assign sigma = 0, log_{e}(0) is undefined and there will be an error))
sigmas = np.linspace(0.01, 1000)

# Calculate entropy for each sigma with list comprehension
entropies = [entropy_normal_distribution(sigma) for sigma in sigmas]

# Plot entropy as function of dependent parameter sigma
plt.figure(figsize=(10, 8))
plt.plot(sigmas, entropies, label='Entropy') # main label showing main function of dependent sigma to entropy
plt.xlabel('Standard Deviation (sigma)')
plt.ylabel('Entropy')
plt.title('Entropy of Normal Distribution and Standard Deviation')
plt.grid(True) # for better understanding limits, let's draw grids()
plt.legend()

# Analyze limits
plt.axvline(x=0.01, color='r', linestyle='--', label='Limit as sigma -> 0.01')
plt.axvline(x=1000, color='y', linestyle='--', label='Limit as sigma -> 1000')

plt.legend()
plt.show()
````

# What is the limit of the function as $\sigma \to +0$?
$$
\lim_{\sigma \to 0^+}(\ln (\sigma \sqrt{2\pi e})) 
\\
= \ln(\lim_{\sigma \to 0^+}(\sigma \sqrt{2\pi e}))
\\
= \ln( \sqrt{2\pi e} \lim_{\sigma \to 0^+}(\sigma)) , \quad lim_{\sigma \to 0^+} f(\sigma) = 0 (\text{Calculus property})
\\
= \ln ( \sqrt{2\pi e} \cdot 0)
\\
= \ln(0) = -\infty
$$
# What is the limit of the function as $\sigma \to +\infty$?
$$
\lim_{\sigma \to +\infty}(\ln (\sigma \sqrt{2\pi e})) 
\\
= \ln(\lim_{\sigma \to +\infty}(\sigma \sqrt{2\pi e}))
\\
= \ln( \sqrt{2\pi e} \lim_{\sigma \to +\infty}(\sigma)),  \quad lim_{\sigma \to +\infty} f(\sigma) = +\infty (\text{sigma is positive and at the numerator})
\\
= \ln ( \sqrt{2\pi e} \cdot +\infty)
\\
= \ln(+\infty) = +\infty
$$

# Finding KL divergence between $p\sim \mathrm{Geom}(s)$ and $q\sim \mathrm{Geom}(t)$, $0 < s, t < 1$:
KullbackLeibler divergence measures distance between two probability distributions. KL divergence is calculated as:
$$
\mathbb{KL}(p\vert\vert q) = \sum\limits_k p_k\log\frac{p_k}{q_k}
$$
Geometric distribution definition:
A discrete random variable $X$ has a geometric distribution with parameter $p$ if it takes on values 1, 2, ..., m, ...(infinite countable set of values) with probabilities:
$$
P(X=m) = pq^{m-1}
$$
where **0<p<1, q=1-p** 
****
In our case, we have two geometric distributions $p, q$ with parameters $s, t$ respectively, their probability mass functions (PMF)s are given by:
$$
p(k) = (1-s)^{k-1}s
\\
q(k) = (1-t)^{k-1}t
$$ 
where $s$ and $t$ are success probabilities for distributions $p$ and $q$ respectively.
Substituting PMFs of geometric distributions into the KL divergence formula:
$$
\mathbb{KL}(p\vert\vert q) = \sum\limits_k p_k\log\frac{p_k}{q_k}
\\
\mathbb{KL}(p\vert\vert q) = \sum\limits_{k=1}^{+\infty}  (1-s)^{k-1}s \log\frac{(1-s)^{k-1}s}{(1-t)^{k-1}t} \quad (1)

$$

$$
Since, sum of all events of probabilities = 1(sum of probabilities of geometric distribution forms series, more preciesly, geometric series, probabilities pi form the geometric progression with the
first member p and denominator q (therefore, the law is said to be geometric):
$$
\sum\limits_{k=1}^{+\infty}  (1-s)^{k-1}s  = 1
$$
Finally KL divergence for geometric series:
$$
\mathbb{KL}(p\vert\vert q) = \frac{1}{1-s} \cdot s \cdot \ln(\frac{1-s}{1-t}) \cdot (\frac{-(1-s) \ln (1-s) + (1-s)}{\ln^2(1-s)}) + \ln (\frac{1-t}{1-s}) + \ln (\frac {s}{t})
$$

# Plotting the graphs of $\mathbb{KL}(p, q)$ as functions of $s$ for several fixed values of $t$.

````python

import numpy as np
import matplotlib.pyplot as plt

def kl_divergence(s, t):
   return ((1-s)/s * np.log((1-s)/(1-t)) + np.log(s/t))
   # return (((s/ (1-s)) * np.log((1-s) / (1-t)))* ((s-1) * np.log(1-s) + 1 - s) / (np.log(1-s) **2) + np.log((1 - t) / (1 - s)) + np.log(s / t))

# Checkiing KL divergence for s = t
s = 0.5
t = 0.5
kl_st = kl_divergence(s, t)
print("KL divergence for s = t:", kl_st)

# Checking if KL(p, q) = KL(q, p)
kl_pq = kl_divergence(0.4, 0.6)
kl_qp = kl_divergence(0.6, 0.4)

print("KL divergence from p to q:", kl_pq)
print("KL divergence from q to p:", kl_qp)

# Plotting KL divergence as function of s for some values of t
t_values = [0.2, 0.5, 0.8]
s_values = np.linspace(0.01, 0.99, 100)
plt.figure(figsize=(10, 6))
for t in t_values:
    kl_values = [kl_divergence(s, t) for s in s_values]
    plt.plot(s_values, kl_values, label=f"t={t}")
plt.xlabel('s')
plt.ylabel('KL divergence')
plt.title('KL Divergence between Geometric Distributions')
plt.legend()
plt.grid(True)
plt.show()
````
